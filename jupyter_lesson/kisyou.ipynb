```python
import requests
response = requests.get('https://www.jma.go.jp/bosai/forecast/data/forecast/130000.json')
```

```python
data = response.json()
data
```

```python
from pprint import pprint

for row in data:
    for key,value in row.items():
        pprint(f"{key}:{value}")
```

```python
areas = []
waves = []
weathers = []
winds = []
for row in data:
    for ts in row["timeSeries"]:
        for area_info in ts["areas"]:
            print(area_info)
            name = area_info["area"]["name"]
            wave = area_info.get("waves")
            weat = area_info.get("weathers")
            wind = area_info.get("winds")
            areas.append(name)
            waves.append(wave)
            weathers.append(weat)
            winds.append(wind)
print(areas)
```

```python
import pandas as pd
df = pd.DataFrame({'name': areas,
                   'waves':waves,
                   'weathers':weathers,
                  'winds':winds})
print(df)
df.to_csv('name.csv',index = False)
```
