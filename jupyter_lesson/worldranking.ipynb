```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import pandas as pd
response = requests.get('https://takkyu-navi.jp/player/ranking')

soup = BeautifulSoup(response.text, "html.parser")

soup
```

```python
title = []
manrank_s = []
womanrank_s =[]
manrank_d = []
womanrank_d =[]
mix_d = []
divs = soup.select('div.col-xs-12.col-sm-12.col-md-6.col-lg-6')
for div in divs:
    title = div.select_one('h2 span')
    if title and title.text.strip() == "男子総合":
        rows = div.select('table.rankingTable tr')
        for row in rows:
            name = row.select_one('span.nameTxt')
            manrank_s.append(name)
            
    if title and title.text.strip() == "女子総合":
        rows = div.select('table.rankingTable tr').text
        for row in rows:
            name = row.select_one('span.nameTxt').text
            womanrank_s.append(name)
            
    if title and title.text.strip() == "男子ダブルス":
        rows = div.select('table.rankingTable tr')
        for row in rows:
            name = row.select_one('span.nameTxt').text
            manrank_d.append(name)
            
    if title and title.text.strip() == "女子ダブルス":
        rows = div.select('table.rankingTable tr')
        for row in rows:
            name = row.select_one('span.nameTxt').text
            womanrank_d.append(name)
            
    if title and title.text.strip() == "混合ダブルス":
        rows = div.select('table.rankingTable tr')
        for row in rows:
            name = row.select_one('span.nameTxt').text
            mix_d.append(name)
    
         #print(name)
print(manrank_s)
print(womanrank_s)
print(manrank_d)
print(womanrank_d)
print(mix_d)


"""
for s in name:
    na = s.select('span')
    title.append(na)
print(name)
#pprint(title)
"""
```

```python
divs = soup.select('div.col-xs-12.col-sm-12.col-md-6.col-lg-6')
for div in divs:
    title_tag = div.select_one('h2 span')
    if not title_tag:
        continue
    title = title_tag.text.strip()

    rows = div.select('table.rankingTable tr')
    for row in rows:
        name_tag = row.select_one('span.nameTxt')
        if not name_tag:
            continue
        name = name_tag.text.strip()

        if title == "男子総合":
            manrank_s.append(name)
        elif title == "女子総合":
            womanrank_s.append(name)
        elif title == "男子ダブルス":
            manrank_d.append(name)
        elif title == "女子ダブルス":
            womanrank_d.append(name)
        elif title == "混合ダブルス":
            mix_d.append(name)
            
print(manrank_s)
print(womanrank_s)
print(manrank_d)
print(womanrank_d)
print(mix_d)
```

```python
divs = soup.select('div.col-xs-12.col-sm-12.col-md-6.col-lg-6')
for div in divs:
    title_tag = div.select_one('h2 span')
    if not title_tag:
        continue
    title = title_tag.text.strip()

    rows = div.select('table.rankingTable tr')
    for row in rows:
        name_tags = row.select('span.nameTxt')
        if not name_tags:
            continue
        names = [n.text.strip() for n in name_tags]

        # 分類
        if title == "男子総合":
            manrank_s.append(names[0])  # シングルスは1人
        elif title == "女子総合":
            womanrank_s.append(names[0])
        elif title == "男子ダブルス":
            manrank_d.append("／".join(names))  # ダブルスは2人
        elif title == "女子ダブルス":
            womanrank_d.append("／".join(names))
        elif title == "混合ダブルス":
            mix_d.append("／".join(names))
print(manrank_s)
print(womanrank_s)
print(manrank_d)
print(womanrank_d)
print(mix_d)
```
